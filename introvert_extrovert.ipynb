```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for x in train.columns:
  if train[x].dtype == 'object':
    train[x].fillna(method='ffill',inplace=True)
    train[x]=le.fit_transform(train[x])
  else:
    train[x].fillna(train[x].mean(),inplace=True)
```

```python
for x in test.columns:
  if test[x].dtype == 'object':
    test[x].fillna(method='ffill',inplace=True)
    test[x]=le.fit_transform(test[x])
  else:
    test[x].fillna(test[x].mean(),inplace=True)
```

```python
x=train.loc[:,['Time_spent_Alone', 'Stage_fear', 'Social_event_attendance',
       'Going_outside', 'Drained_after_socializing', 'Friends_circle_size',
       'Post_frequency']]
y=train['Personality']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression()
DTC=DecisionTreeClassifier()
```

```python
LR.fit(x,y)
```

```python
DTC.fit(x,y)
```

```python
print(LR.score(x,y))
print(DTC.score(x,y))
```

```python
test.sample()
```

```python
test_x = test.drop('id',axis=1)
```

```python
Lr_pred=LR.predict(test_x)
DTC_pred=DTC.predict(test_x)
```

```python
print(LR.score(test_x,Lr_pred))
print(DTC.score(test_x,DTC_pred))
```
