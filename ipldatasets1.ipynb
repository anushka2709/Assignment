```python
#most runs by a batsman
df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
#most wickets by a bowler
df.groupby('bowler')['player_dismissed'].sum().sort_values(ascending=False)
```

```python
#most sixes hit
df[df['batsman_runs']==6]['batsman'].value_counts()
```

```python
#most dot balls
dotballs = df[df['total_runs']==0]
dotballs['bowler'].value_counts()
```

```python
#most runs by a team
df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
```

```python
#highest run made by a batsman in a match
highscore = df.groupby(['match_id', 'batsman'])['batsman_runs'].sum().reset_index()
batsmanscoringhighest = highscore.sort_values(by='batsman_runs', ascending=False)
print(batsmanscoringhighest)
```
